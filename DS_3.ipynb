### 3.1 matplotlib

```python
from matplotlib import pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# x축 : 연도, y축 : GDP
plt.plot(years, gdp, color = 'green', marker ='o', linestyle = 'solid')

# 제목
plt.title("Nominal GDP")

# y축 label
plt.ylabel("Billions of $")
plt.show()
```

### 3.2 막대 그래프 : 이산적인(discrete) 항목들에 대한 변화 시각화 할 때 사용


```python
movies = ["Annie Hall", "Ben-Hur", "Casablanca", "Gandhi", "West Side Story"]
num_oscars = [5, 11, 3, 8, 10]

# x축 : [0, 1, 2, 3, 4], y축 : [num_oscar]
plt.bar(range(len(movies)), num_oscars)

plt.title("My Favorite Movies")
plt.ylabel("# of Academy Awards")

# x축 : 영화 제목 추가
plt.xticks(range(len(movies)), movies)

plt.show()
```

```python
# Histogram : 정해진 구간에 해당되는 항목의 개수를 통해 분포를 관찰 가능

from collections import Counter 
grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# 점수는 10점 단위로 그룹화, 100점은 90점대!
histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5for x in histogram.keys()], # 막대를 5만큼씩 오른쪽으로 이동
        histogram.values(), # 막대 높이
        10, # 너비 : 10
        edgecolor = (0, 0, 0)) # 막대 테두리 : 검은색

plt.axis([-5, 105, 0, 5]) # x축 : -5 ~ 105 / y축 : 0 ~ 5 => 0부터 시작 안하면 밑에 plot처럼 그려짐.

plt.xticks([10 * i for i in range(11)]) # x축 레이블 : 0, 10, ..., 100
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

### 3.3 선 그래프 : 경향 파악할 때 유용

```python
variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

# 한 plot에 여러 선
plt.plot(xs, variance, 'g-', label = 'variance') # 실선
plt.plot(xs, bias_squared, 'r-', label = 'bias^2') # 일점쇄선
plt.plot(xs, total_error, 'b:', label = 'total error') # 점선

# 범례
plt.legend(loc = 9)
plt.xlabel("model complexity")
plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

### 3.4 산점도 : 두 변수간의 연관 관계 파악

```python
friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)

# 각 point에 label 달기
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, xy = (friend_count, minute_count), # label 
                xytext = (5, -5), # label 약간 point에서 떨어져 있게 하기
                textcoords = 'offset points')
    
plt.title("Daily Minutes vs, Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("daily minutes spent on this site")
plt.show()
```

```python
# 변수들끼리 비교할 때 자동으로 축의 범위를 설정하면 밑의 plot처럼 공정한 비교를 못하게 됤 수도 있음.
test_1_grades = [99, 90, 85, 97, 80]
test_2_grades = [100, 85, 60, 90, 70]

plt.scatter(test_1_grades, test_2_grades)
plt.title("Axes Aren't Comparable")
plt.xlabel("test 1 grade")
plt.ylabel("test 2 grade")
plt.axis("equal") # 추가하면 공정한 비교 가능
plt.show()


```
